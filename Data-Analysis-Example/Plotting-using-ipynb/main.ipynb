# Violin Plot

In general, violin plots are mnethod of plotting numeric data and can be considered a combination of the box with a kernel density plot.
In the violin, we can find the same information as in the box plots:
- median: a white dot in the violin plot
- interquartile range: the black bar in the center of violin
- the adjacent values: the black lines stretched from the bar which defined as `first quartile - 1.5 IQR` and `third quartile + 1.5 IQR` respectively. 

Th unquestionable advantage of the violin plot over the box plot is that, aside from showing the abovementioned statistics, it also shows the entire distribution of the data. This is interest, particularly when dealing with multimodal data, i.e. a distribution with more than one peak.

# Data Pipeline 

```
# create empty dataframe
df = pd.DataFrame()

# create a column
df["name"] = ["alpha", "bravo", "charlie"]
df["employed"] = ["Yes", "Yes", "No"]
df["age"] = [32, 29, 19]

# review the dataframe
df
```

```
# create a function
def mean_age_by_group(dataframe, col):
  # return the mean value of age per group
  return dataframe.groupby(col).mean()
```

```
# cerate a function
def uppercase_column_name(dataframe):
  # caps all the column headers
  dataframe.columns = dataframe.columns.str.upper()
  # return the data with modified column header
  return dataframe
```

```
# create a pipeline that applies both functions simultaniously
(df.pipe(mean_age_by_group, col="employed")
   .pipe(uppercase_column_name)
)
```
